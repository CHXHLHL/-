## 条形图(barh)

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax1=fig.subplots()
ax2=ax1.twiny()#与twinx相反
name = ['张三','李四','王五','胡六']
chinese = [85,96,74,62]
math = [89,98,62,79]
#bar改为barh,width改为height,透明度为0.8,从0到1透明度递增,bar也有此属性
ax1.barh(name,chinese,height=-0.2,color=['r'],
        linewidth=2,edgecolor='black',align='edge',label='语文',alpha=0.8)
ax2.barh(name,math,height=0.2,color=['g'],
        linewidth=2,edgecolor='black',align='edge',label='数学')
ax1.set_xlabel('语文')#设置x轴名称
ax2.set_xlabel('数学')
fig.legend(loc='upper center',bbox_to_anchor=(0.75,0.8))
plt.title('学生成绩',position=(0.5,1.2))#更改题目名称
plt.show()
```

## 折线图

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np 
x = np.linspace(-2*np.pi,2*np.pi,100)
y = np.cos(x)
plt.plot(x,y)#线条默认为实线
plt.show()
```

```python
#设置线条颜色，线条样式，标记样式，线条宽度，标记大小
plt.plot(x,y,color='g',linestyle='--',marker='o',linewidth=1,markersize=5)
plt.show()
```

支持的颜色名称简写有如下几种

'r':红色          'y':黄色

'b':蓝色          'g':绿色

'c':青色          'k':黑色

'w':白色          'm':品红

支持的线条样式

'-'或'solid'(默认):实线

'--'或'dashed':虚线

'-.'或'dashdot':杠点线

':'或'dotted':点线

''或' '或'None':无

```python
#通过合并简写线条颜色，线条样式和标记样式,markersize可简写为ms
plt.plot(x,y,'g--o',linewidth=1,markersize=5)
plt.show()
```

```python
#mec：markeredgecolor的简写,标记轮廓颜色
#mew：markeredgewidth的简写，标记轮廓宽度
#mfc：markerfacecolor的简写，标记颜色
plt.plot(x,y,'g--o',linewidth=1,markersize=5,mec='r',mew=2,mfc='k')
plt.show()
```

```python
y1 = np.cos(x)
y2 = np.sin(x)
plt.plot(x,y1,label='cosx')
plt.plot(x,y2,'r--',label='sinx')
ax = plt.gca()#获取当前子图
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')#将底部坐标作为x轴
ax.spines['bottom'].set_position(('data', 0))#移动x轴至y=0处
ax.yaxis.set_ticks_position('left')#将左边作为y轴
ax.spines['left'].set_position(('data', 0))#移动y轴至x=0处
plt.xlabel('x',fontsize=20)#设置x坐标名称以及大小
ax.xaxis.set_label_coords(1,0.5)#设置x坐标名称位置，与bbox_to_anchor基本一致
plt.ylabel('y',rotation='horizontal',fontsize=20)
ax.yaxis.set_label_coords(0.5,1)#设置y坐标名称位置
#标注某点，以下标注内容是(2,0),标注的点是(2,0),标注样式是箭头,箭头起始位置是(4,0.5)
plt.annotate('(2,0)',xy=(2,0), arrowprops=dict(arrowstyle='->'), xytext=(4, 0.5),fontsize=15)
plt.grid()#添加网格
plt.legend()
plt.show()
```

## 饼图

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
labels = ['苹果','香蕉','葡萄','橙子','西瓜']#标签，元素个数需与x的元素个数相同
explode = [0,0.2,0,0,0]#指定偏离圆心的距离，类型也可以是元组，元素个数需与x的元素个数相同
plt.pie(x,labels=labels,explode=explode,shadow=True)
plt.axis('equal')#为了确保是一个圆
plt.show()
```

```python
#保留一位小数并以百分比的形式展现，开始角度为90°(以最大份额的饼图块为准),字体为黑色,标签旋转90°
plt.pie(x,labels=labels,explode=explode,shadow=True,autopct='%.1f%%',
        startangle=90,textprops=dict(color='k'),rotatelabels=90)
plt.legend(bbox_to_anchor=(1,0.7),title='水果种类')
plt.axis('equal')
plt.show()
```

```python
#修改标签到圆心的距离(默认是1.1)和值到圆心的距离(默认是0.6),两者均为比值而不是真正的距离
plt.pie(x,labels=labels,labeldistance=0.5,autopct='%.1f%%',pctdistance=1)
plt.axis('equal')
plt.show()
```

```python
#设置半径的大小，虽然和上图看起来大小一样，但是通过显示坐标轴看到半径的确增大了
plt.pie(x,radius=2,frame=True)
plt.axis('equal')
plt.show()
```

```python
outer = [1,2,3,4]
inner = [1,2,3,4,5,6,7]
plt.pie(outer,wedgeprops=dict(width=0.3,edgecolor='w'))
plt.pie(inner,radius=0.7,wedgeprops=dict(width=0.2,edgecolor='w'))
plt.axis('equal')
plt.title('甜甜圈')
plt.show()
```

## 3D图

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)#创建三维坐标系
x = np.random.randint(0,10,50)
y = np.random.randint(0,10,50)
z = np.random.randint(0,10,50)
ax.scatter(x, y, z, s=100, c=x,cmap='hsv')
ax.set_xlabel('x轴')
ax.set_ylabel('y轴')
ax.set_zlabel('z轴')#此处不可以使用plt.zlabel()
ax = plt.gca()
ax.patch.set_facecolor('red')#设置背景颜色
ax.patch.set_alpha(0.5)#设置透明度
plt.show()
```

```python
fig1 = plt.figure()
ax1 = Axes3D(fig1)
theta = np.linspace(-8*np.pi, 8 * np.pi, 400)
z = np.linspace(-4, 4, 400)
r = 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax1.plot(x, y, z)
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.show()
```

```python
%matplotlib inline
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
fig2 = plt.figure()
ax2 = Axes3D(fig2)
center = [0,0,0]#球心坐标
radius = 10#半径
phi = np.linspace(0, 2 * np.pi, 100)#φ角
theta = np.linspace(0, np.pi, 100)#θ角
x = radius * np.outer(np.cos(phi), np.sin(theta)) + center[0]#x=r*cosφ*sinθ
y = radius * np.outer(np.sin(phi), np.sin(theta)) + center[1]#y=r*sinφ*sinθ
z = radius * np.outer(np.ones(100), np.cos(theta)) + center[2]#z=r*cosθ
surf = ax2.plot_surface(x,y,z,cmap=cm.rainbow,alpha=0.8)
ax2.contourf(x,y,z,zdir='x',offset=-10)#在yz面上的等高线
ax2.contourf(x,y,z,zdir='y',offset=10)
ax2.contourf(x,y,z,zdir='z',offset=-10)
ax2.set_zlim(-9,9)#为了看起来更像球
fig2.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
